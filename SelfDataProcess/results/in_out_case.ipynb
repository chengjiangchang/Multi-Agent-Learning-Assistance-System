```python
import pandas as pd
from datetime import datetime

# è¯»å–æ•°æ®
tutoring_df = pd.read_csv("tutoring_content_results_qwen-plus.csv")
mastery_df = pd.read_csv("mastery_assessment_results_minimal_qwen-plus.csv")
assessment_df = pd.read_pickle("experiment_results_both_qwen-plus.pkl")

# è·å–tutoringä¸­çš„å­¦ç”ŸIDå¹¶ç­›é€‰
tutoring_student_ids = set(tutoring_df['student_id'].unique())
filtered_mastery_df = mastery_df[mastery_df['student_id'].isin(tutoring_student_ids)].copy()
filtered_assessment_df = assessment_df[assessment_df['student_id'].isin(tutoring_student_ids)].copy()

# æŒ‰å­¦ç”ŸIDå’ŒçŸ¥è¯†ç‚¹æ’åº
tutoring_df_sorted = tutoring_df.sort_values(['student_id', 'kc_name']).reset_index(drop=True)
filtered_mastery_df_sorted = filtered_mastery_df.sort_values(['student_id', 'kc_name']).reset_index(drop=True)
filtered_assessment_df_sorted = filtered_assessment_df.sort_values(['student_id', 'question_id']).reset_index(drop=True)

# ç”Ÿæˆtxtæ–‡ä»¶
output_file = "llm_prompts_and_responses.txt"

with open(output_file, 'w', encoding='utf-8') as f:
    # å†™å…¥æ–‡ä»¶å¤´
    f.write("=" * 100 + "\n")
    f.write("LLM PROMPTS AND RESPONSES - TUTORING & MASTERY ASSESSMENT (BY KNOWLEDGE CONCEPT)\n")
    f.write(f"Generated at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write("=" * 100 + "\n\n")
    
    # ç»Ÿè®¡ä¿¡æ¯
    f.write(f"Total Students: {len(tutoring_student_ids)}\n")
    f.write(f"Tutoring Records: {len(tutoring_df_sorted)}\n")
    f.write(f"Mastery Assessment Records: {len(filtered_mastery_df_sorted)}\n")
    f.write(f"Assessment Agent Records: {len(filtered_assessment_df_sorted)}\n")
    f.write("=" * 100 + "\n\n\n")
    
    # éå†æ¯ä¸ªå­¦ç”Ÿ
    for student_id in sorted(tutoring_student_ids):
        f.write("#" * 100 + "\n")
        f.write(f"STUDENT ID: {student_id}\n")
        f.write("#" * 100 + "\n\n")
        
        # è·å–è¯¥å­¦ç”Ÿçš„æ‰€æœ‰è®°å½•
        student_tutoring = tutoring_df_sorted[tutoring_df_sorted['student_id'] == student_id]
        student_mastery = filtered_mastery_df_sorted[filtered_mastery_df_sorted['student_id'] == student_id]
        student_assessment = filtered_assessment_df_sorted[filtered_assessment_df_sorted['student_id'] == student_id]
        
        # è·å–è¯¥å­¦ç”Ÿæ¶‰åŠçš„æ‰€æœ‰çŸ¥è¯†ç‚¹ï¼ˆå–ä¸¤è€…çš„å¹¶é›†ï¼‰
        tutoring_kcs = set(student_tutoring['kc_name'].unique())
        mastery_kcs = set(student_mastery['kc_name'].unique())
        all_kcs = sorted(tutoring_kcs | mastery_kcs)
        
        kc_counter = 0
        
        # æŒ‰çŸ¥è¯†ç‚¹éå†
        for kc_name in all_kcs:
            kc_counter += 1
            f.write("+" * 100 + "\n")
            f.write(f"KNOWLEDGE CONCEPT #{kc_counter}: {kc_name}\n")
            f.write("+" * 100 + "\n\n")
            
            # ========== è¯¥çŸ¥è¯†ç‚¹çš„ TUTORING CONTENT ==========
            kc_tutoring = student_tutoring[student_tutoring['kc_name'] == kc_name]
            
            if len(kc_tutoring) > 0:
                f.write(">>> TUTORING CONTENT:\n")
                f.write("-" * 100 + "\n")
                
                for idx, row in kc_tutoring.iterrows():
                    f.write("SYSTEM PROMPT:\n")
                    f.write(str(row['prompt_system']) + "\n\n")
                    
                    f.write("USER PROMPT:\n")
                    f.write(str(row['prompt_user']) + "\n\n")
                    
                    f.write("LLM RESPONSE:\n")
                    f.write(str(row['llm_raw_response']) + "\n\n")
                    
                    f.write("EXTRACTED TUTORING CONTENT:\n")
                    f.write(str(row['tutoring_content']) + "\n\n")
                    f.write("-" * 100 + "\n")
            else:
                f.write(">>> TUTORING CONTENT: [No tutoring record for this KC]\n")
                f.write("-" * 100 + "\n")
            
            f.write("\n")
            
            # ========== è¯¥çŸ¥è¯†ç‚¹çš„ MASTERY ASSESSMENT ==========
            kc_mastery = student_mastery[student_mastery['kc_name'] == kc_name]
            
            if len(kc_mastery) > 0:
                f.write(">>> MASTERY ASSESSMENT:\n")
                f.write("-" * 100 + "\n")
                
                for idx, row in kc_mastery.iterrows():
                    f.write(f"MASTERY LEVEL: {row['mastery_level']}\n\n")
                    
                    f.write("SYSTEM PROMPT:\n")
                    f.write(str(row['prompt_system']) + "\n\n")
                    
                    f.write("USER PROMPT:\n")
                    f.write(str(row['prompt_user']) + "\n\n")
                    
                    f.write("LLM RESPONSE:\n")
                    f.write(str(row['llm_raw_response']) + "\n\n")
                    
                    f.write("EXTRACTED FIELDS:\n")
                    f.write(f"  - Mastery Level: {row['mastery_level']}\n")
                    f.write(f"  - Rationale: {row['rationale']}\n")
                    f.write(f"  - Suggestions: {row['suggestions']}\n\n")
                    f.write("-" * 100 + "\n")
            else:
                f.write(">>> MASTERY ASSESSMENT: [No mastery record for this KC]\n")
                f.write("-" * 100 + "\n")
            
            f.write("\n\n")
        
        # ========== è¯¥å­¦ç”Ÿçš„ ASSESSMENT AGENT (æµ‹è¯„æ™ºèƒ½ä½“) ==========
        if len(student_assessment) > 0:
            f.write("+" * 100 + "\n")
            f.write("ASSESSMENT AGENT (æµ‹è¯„æ™ºèƒ½ä½“)\n")
            f.write("+" * 100 + "\n\n")
            
            question_counter = 0
            for idx, row in student_assessment.iterrows():
                question_counter += 1
                f.write(">>> QUESTION #{}: (ID: {})\n".format(question_counter, row['question_id']))
                f.write("-" * 100 + "\n")
                
                f.write(f"KNOWLEDGE CONCEPT: {row['true_know_name']}\n")
                f.write(f"TRUE SCORE: {row['true_score']}\n")
                f.write(f"TRUE ANSWER CHOICE ID: {row['true_answer_choice_id']}\n")
                f.write(f"EXPERIMENT TYPE: {row['experiment_type']}\n")
                f.write(f"EXPERIMENT MODE: {row['experiment_mode']}\n\n")
                
                f.write("SYSTEM PROMPT:\n")
                f.write(str(row['prompt_system']) + "\n\n")
                
                f.write("USER PROMPT:\n")
                f.write(str(row['prompt_user']) + "\n\n")
                
                f.write("LLM RESPONSE:\n")
                f.write(str(row['llm_raw_response']) + "\n\n")
                
                f.write("PREDICTED RESULTS:\n")
                f.write(f"  - Task1 (Self Predict): {row['predicted_task1_selfpredict']}\n")
                f.write(f"  - Task2 (Know Name): {row['predicted_task2_know_name']}\n")
                f.write(f"  - Task3 (Reasoning): {row['predicted_task3_reasoning']}\n")
                f.write(f"  - Task4 (Answer Choice): {row['predicted_task4_answer_choice']}\n\n")
                
                f.write("-" * 100 + "\n\n")
            
            f.write("\n")
        else:
            f.write("+" * 100 + "\n")
            f.write("ASSESSMENT AGENT (æµ‹è¯„æ™ºèƒ½ä½“): [No assessment records for this student]\n")
            f.write("+" * 100 + "\n\n")
        
        f.write("=" * 100 + "\n")
        f.write(f"END OF STUDENT {student_id}\n")
        f.write("=" * 100 + "\n\n\n")

print(f"âœ… æˆåŠŸç”Ÿæˆæ–‡ä»¶: {output_file}")
print(f"\næ–‡ä»¶ç»Ÿè®¡:")
print(f"  - å­¦ç”Ÿæ•°é‡: {len(tutoring_student_ids)}")
print(f"  - Tutoringè®°å½•æ•°: {len(tutoring_df_sorted)}")
print(f"  - Masteryè®°å½•æ•°: {len(filtered_mastery_df_sorted)}")
print(f"  - Assessmentè®°å½•æ•°: {len(filtered_assessment_df_sorted)}")
print(f"\nå‰5ä¸ªå­¦ç”ŸID: {sorted(tutoring_student_ids)[:5]}")
```
